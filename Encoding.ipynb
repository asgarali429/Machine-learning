###Encoding a Categorical Feature

```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.head()
```

```python
df = df.iloc[:,2:]
df
```

##Ordinal Encoding
1.Manually

```python
df['review']=df['review'].map({'Poor':0,'Average':1,'Good':2})
df['education']=df['education'].map({'School':0,'UG':1,'PG':2})
```

```python
df.head()
```

2.Using scitlearn OrdinalEncoder

```python
df1 = pd.read_csv('customer.csv')
```

```python
df1=df1.iloc[:,2:]
df1
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(df1.iloc[:,:2],df.iloc[:,-1],test_size=0.2)
```

```python
xtrain = oe.fit_transform(xtrain)
```

```python
xtest = oe.transform(xtest)
```

```python
xtrain
```

```python
xtest
```

3.Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
le.fit(ytrain)

ytrain= le.transform(ytrain)
ytest = le.transform(ytest)
```

```python
ytrain
```

we use a ordinal and label encoding when a categories are ordinal .we apply ordinal encoding in x feature(independent) and label encoding apply in y or label(dependent) feature

```python
cars['fuel'].value_counts()
```

4. OneHotEncoding with Top Categories

```python

counts = cars['brand'].value_counts()

cars['brand'].nunique()
threshold = 100

repl = counts[counts <= threshold].index

pd.get_dummies(cars['brand'].replace(repl, 'uncommon')).sample(5)
```

we use OneHotEncoding for Nominal data
